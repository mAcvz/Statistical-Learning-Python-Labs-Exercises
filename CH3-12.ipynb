# Exercise 12 — Equal Coefficients in No-Intercept Regression

This exercise explores the conditions under which swapping predictor and response 
in a simple linear regression without intercept yields the same coefficient estimate.

* Determine the condition under which the coefficient estimate for the 
          regression of $X$ onto $Y$ equals the coefficient estimate for the 
          regression of $Y$ onto $X$, both without intercept.
    
* Generate a synthetic dataset with $n = 100$ observations where the 
          two coefficient estimates are \textbf{different}.
    
* Generate a synthetic dataset with $n = 100$ observations where the 


$$ \hat{\beta}_{Y \text{ onto } X} = \frac{\sum_{i=1}^n x_i y_i}{\sum_{i=1}^n x_i^2}, \qquad
\hat{\beta}_{X \text{ onto } Y} = \frac{\sum_{i=1}^n x_i y_i}{\sum_{i=1}^n y_i^2} $$

$$\hat{\beta}_{Y \text{ onto } X} = \hat{\beta}_{X \text{ onto } Y} \iff \sum_{i=1}^n x_i^2 = \sum_{i=1}^n y_i^2$$
