```python
import numpy as np
import pandas as pd
from matplotlib.pyplot import subplots

import statsmodels.api as sm

from statsmodels.stats.outliers_influence \
import variance_inflation_factor as VIF
from statsmodels.stats.anova import anova_lm

from ISLP import load_data
from ISLP.models import (ModelSpec as MS ,
                        summarize ,
                        poly)
```

```python
rng = np.random.default_rng(1)
x = rng.normal(size=100)
y = 2 * x + rng.normal(size=100)
```

```python
df1 = pd.DataFrame({'x':x, 'y':y})
X = MS(['x'],intercept=False).fit_transform(df1)
model1 = sm.OLS(y,X)
results1 = model1.fit()
summarize(results1)
```

We would have expect to obtain 2 in a perfect experiment, but we added a random $\epsilon_i$ to each data point so some noise is there. 

```python
model2 = sm.OLS(x,y)
results2 = model2.fit()
summarize(results2)
```

The t-value and p-value are the same because the formula is symmetric in $x$ and $y$:
$$ t_j = \frac{\hat{\beta_j}-\beta_o}{\text{SM}(\hat{\beta_j})} $$

where:

$$ \text{SE}(\hat{\beta}) = \sqrt{\frac{\sum_{i=1}^{n}(y_i - x_i\hat{\beta})^2}{(n-1)\sum_{i=1}^{n}x_{i}^2}} $$
$$t_j = \frac{\hat{\beta_j}}{\text{SM}(\hat{\beta_j})} = \frac{\sqrt{n-1}\,\sum x_i y_i}{\sqrt{(\sum x_i^2)(\sum y_i^2) - (\sum x_i y_i)^2}} $$

```python
# y onto x (with intercept)
model_a = sm.OLS(df1['y'], sm.add_constant(df1['x'])).fit()
print(summarize(model_a))

# x onto y (with intercept)
model_b = sm.OLS(df1['x'], sm.add_constant(df1['y'])).fit()
print(summarize(model_b))
```
